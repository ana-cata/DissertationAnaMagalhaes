```python
A1 = "20210624-143847-peixe1.csv"
A2 = "20210624-145602-peixe2.csv"
A3 = "20210624-151249-peixe3.csv"
A4 = "20210624-153033-peixe4.csv"
A5 = "20210624-155357-peixe5.csv"
data=pd.read_csv(A1)
data.columns = ['Tempo','Amplitude', 'Temperatura']
b = 11
f = 600
print(data)
begin= int(b/0.05)- int(10.45/0.05) 
finish = int(f/0.05)- int(10.45/0.05)
```

```python
plt.plot(data['Tempo'],data['Temperatura'], 'o')

plt.xlabel('Time(s)')
plt.ylabel('Heart beat Reading')
plt.show()
```

```python
z = np.polyfit(data['Tempo'], data['Temperatura'], 1)
p = np.poly1d(z)
thermistor_erro = 0.2
y_min1 = []
y_max1 =[]
for m1 in range(0, len(data['Tempo'])):
    y_min1.append(p(data['Tempo'][m1]) - thermistor_erro)         
    y_max1.append(p(data['Tempo'][m1]) + thermistor_erro)



plt.plot(data['Tempo'],data['Temperatura'], 'o')
# plt.plot(data['Tempo'],p(data['Tempo']),"r-")
plt.fill_between(data['Tempo'], y_min1, y_max1,  color='C0', alpha=0.3)
plt.ylim([20,25])
```
